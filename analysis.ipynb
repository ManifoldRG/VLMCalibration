```python
df1 = pd.read_csv('train_results/records_train_full_gemma.csv', index_col=0)
df2 = pd.read_csv('train_results/records_train_full_qwen.csv', index_col=0)
df3 = pd.read_csv('train_results/records_train_full_llama.csv', index_col=0)

# df1 = pd.read_csv('records_test_full_mid_gemma.csv', index_col=0)
# df2 = pd.read_csv('records_test_full_mid_qwen.csv', index_col=0)


df1
```

```python
# Sort both dataframes by question to align them
df1_sorted = df1.sort_values('question').reset_index(drop=True)
df2_sorted = df2.sort_values('question').reset_index(drop=True)
df3_sorted = df3.sort_values('question').reset_index(drop=True)

# print("First few questions match after sorting:", all(df1_sorted['question'].head() == df2_sorted['question'].head()))
# print("All questions match after sorting:", all(df1_sorted['question'] == df2_sorted['question']))
```

```python
import numpy as np
bins = np.arange(0, 1.1, 0.1)
labels = [f'{bins[i]:.1f}-{bins[i+1]:.1f}' for i in range(len(bins)-1)]

# Get value counts for each bin
df1_counts = pd.cut(df1_sorted['p_true'], bins=bins, labels=labels).value_counts().sort_index()
df2_counts = pd.cut(df2_sorted['p_true'], bins=bins, labels=labels).value_counts().sort_index()
df3_counts = pd.cut(df3_sorted['p_true'], bins=bins, labels=labels).value_counts().sort_index()
# Create comparison dataframe
comparison_df = pd.DataFrame({
    'Gemma': df1_counts,
    'Qwen': df2_counts,
    'Llama': df3_counts
})

print("P_true distribution in bins of 0.1:")
print(comparison_df)
```
